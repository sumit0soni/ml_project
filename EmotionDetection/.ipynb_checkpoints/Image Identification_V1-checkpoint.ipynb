```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np

def preprocess_image(img_path, target_size=(96, 96)):
    # Load the image using Keras
    img = image.load_img(img_path, target_size=target_size)
    
    # Convert the PIL Image to a NumPy array
    img_array = image.img_to_array(img)

    # Add an extra batch dimension (for compatibility with preprocess_input)
    img_array = np.expand_dims(img_array, axis=0)

    # Preprocess the image array for ResNet50
    img_array = preprocess_input(img_array)

    # Remove the extra batch dimension added by preprocess_input
    img_array = img_array[0]

    return img_array

preprocessed_image = preprocess_image(data['pth'][0])
print(preprocessed_image.shape)
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import cv2
import numpy as np

def preprocess_image(img_path, target_size=(96, 96), grayscale=True):
    # Load the image
    img = image.load_img(img_path, target_size=target_size, color_mode='grayscale' if grayscale else 'rgb')

    # Convert the image to a NumPy array
    img_array = image.img_to_array(img)

    # Preprocess the image array
    img_array = preprocess_input(img_array)

    return img_array
```

```python
from tensorflow.keras.utils import plot_model
from tensorflow.keras import models, layers
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten

# create model structure
model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(96, 96, 1)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

# Plot the model
plot_model(model, to_file='/Users/sumitsoni/Desktop/data_science/project_identify_reaction/model_plot.png', show_shapes=True, show_layer_names=True)
model.summary()
```

```python
num_epochs=10
batch_size=64
history = model.fit(X_train_preprocessed, y_train_encoded, epochs=num_epochs, batch_size=batch_size, validation_split=0.2)
```
